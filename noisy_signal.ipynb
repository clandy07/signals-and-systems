```python
import numpy as np
import matplotlib.pyplot as plt
```

Signal Parameters (from the example)

```python
sampling_rate = 1000  # Hz
T = 1.0  # seconds
N = int(sampling_rate * T)
t = np.linspace(0, T, N, endpoint=False)
```

Signal with two frequencies: 50 Hz and 120 Hz (from the example)

```python
freq1 = 50
freq2 = 120
signal = np.sin(freq1 * 2 * np.pi * t) + 0.5 * np.sin(freq2 * 2 * np.pi * t)
```

Step 1: Add random noise (using np.random.randn(); scaling by 0.2 for visibility of peaks)


```python
noise = 0.2 * np.random.randn(N)
noisy_signal = signal + noise
```

Step 2: Compute the FFT

```python
noisy_fft = np.fft.fft(noisy_signal)
noisy_magnitude = 2.0 / N * np.abs(noisy_fft[:N//2])
freq_axis = np.fft.fftfreq(N, 1 / sampling_rate)[:N//2]
```

Step 2: Plot the results

```python
plt.figure(figsize=(12, 5))
```

Plot the noisy signal in time domain


```python
plt.subplot(1, 2, 1)
plt.plot(t, noisy_signal)
plt.title('Noisy Signal in Time Domain')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

Plot the frequency spectrum

```python
plt.subplot(1, 2, 2)
plt.plot(freq_axis, noisy_magnitude)
plt.title('Noisy Signal Spectrum (via FFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 200)  # Zoom in to see relevant frequencies

plt.show()
```
