Function to implement the echo system y[n] = x[n] + 0.5 * x[n-4]

```python
def echo_system(x):
    y = []
    for n in range(len(x)):
        if n < 4:
            y.append(x[n])  # No echo for first 4 samples
        else:
            y.append(x[n] + 0.5 * x[n - 4])
    return y
```

Example usage

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]  # Sample input signal
y = echo_system(x)
print("Input:", x)
print("Output:", y)
```

2. Classify the system based on the five properties we discussed. Justify your answers.
- Linearity: The system is linear. The equation y[n] = x[n] + 0.5 * x[n-4] follows superposition.
   For inputs a*x1[n] + b*x2[n], output is a*y1[n] + b*y2[n], confirming linearity.
- Time-Invariance: The system is time-invariant. A time shift in input shifts the output identically.
- Causality: The system is causal. Output y[n] depends on x[n] and past x[n-4], not future values.
- Memory: The system has memory. It uses x[n-4], indicating dependence on past input.
- Stability: The system is stable. With a scaling factor of 0.5, the output remains bounded for bounded input.

3. What kind of system is this? (Hint: Is it an LTI system?)

  This is an LTI (Linear Time-Invariant) system. It satisfies linearity and time-invariance, and is also causal and stable, typical of LTI systems in signal processing like this echo.

Visualization


```python
plt.figure(figsize=(10, 6))
plt.plot(x, label='Input Signal x[n]')
plt.plot(y, label='Output Signal y[n]', linestyle='--')
plt.xlabel('Sample n')
plt.ylabel('Amplitude')
plt.title('Echo System: Input vs Output')
plt.legend()
plt.grid(True)
plt.show()
```
